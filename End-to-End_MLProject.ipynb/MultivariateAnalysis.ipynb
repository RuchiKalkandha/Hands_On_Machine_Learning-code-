```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
titanic = pd.read_csv('train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
#scatter plot (numerical - numerical) columns
sns.scatterplot(data=tips, x='total_bill', y='tip',hue=tips['sex'],style = tips['smoker'],size=tips['size'])
plt.show()
```

```python
# bar-plot (numerical - categorical) columns
sns.barplot(titanic, x = titanic['Pclass'], y = titanic['Age'])
```

```python
sns.barplot(titanic, x = titanic['Pclass'], y = titanic['Fare'],hue=titanic['Sex'])
```

```python
# box-plot (numerical-categorical)
sns.boxplot(titanic, x = titanic['Sex'], y = titanic["Age"], hue = titanic['Survived'])
```
