```python
from sklearn.datasets import make_regression
import numpy as np
X,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets=1, noise = 80, random_state=13)
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# as we have got m and b, we will now code gradient descent
# assuming that we know m and need to find b
# slope = constant = 78.35
# lets start with a random value of b = 0
y_pred = ((78.35*X)+0).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slope = -2*np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Let's take learning rate as 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept
b = b- step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color = 'black', label = 'OLS')
plt.plot(X,y_pred1,color='#00a65a',label= 'b= {}'.format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
#  Iteration 2
loss_slope = -2*np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color = 'black', label = 'OLS')
plt.plot(X,y_pred2,color='#00a65a',label= 'b= {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label= 'b= {}'.format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
#  Iteration 3
loss_slope = -2*np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color = 'black', label = 'OLS')
plt.plot(X,y_pred3,color='#00a65a',label= 'b= {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label= 'b= {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label= 'b= {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# Using loops rather than manually writing each iteration
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2*np.sum(y - m*X.ravel() - b)
    b = b - (lr*loss_slope)
    print(b)
    y_pred = m * X + b
    plt.plot(X, y_pred)

plt.scatter(X,y)
```
